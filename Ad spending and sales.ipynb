How would we know that ad spending increases sales?

Well it would be a good idea to conduct an experiment where,
we take a region and display ads for x product
and then take another region where we won't display ads for that product.

Then we compare the two scenarios by creating 2 plots,
in which the x axis is the number of sales
and y axis is ad spending

Consider the following plot to represent a lineplot of our ad fueled region for our given product

```python
from sympy.abc import s
>>> from sympy.physics.control.lti import TransferFunction
>>> from sympy.physics.control.control_plots import ramp_response_plot
>>> tf1 = TransferFunction(s, (s+6)*(s+8), s)
>>> ramp_response_plot(tf1, upper_limit=2)   
```

Since y axis is ad spending, x axis is number of sales of our product, we can clearly see that ad spending 
does have an effect on sales. 
But of course up to a point! We see that after a given spending sum, our sales fail to rise as anticipated.

We have reached the point of diminishing returns.

Proving hypothesis again by removing ad spending

Say we heard that a local influencer of that region,
happened to have started advertising the delightful product we sell so successfully.
Therefore we speculate, that the rise in sales in that region is due to the influencer's reach, instead of our ads.
We can remove ad spending, and since the influencer kept talking about the product, 
we can see whether theres an effect on sales.

Consider the following plot to represent y axis as ad spending, x axis is number of sales

```python
from sympy.abc import s
>>> from sympy.physics.control.lti import TransferFunction
>>> from sympy.physics.control.control_plots import impulse_response_plot
>>> tf1 = TransferFunction(8*s**2 + 18*s + 32, s**3 + 6*s**2 + 14*s + 24, s)
>>> impulse_response_plot(tf1)    
```

The influencer doesn't have much influence after all!
